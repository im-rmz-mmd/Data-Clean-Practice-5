```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tour.csv').set_index('Rank')
df
```

```python
df = df.drop(['Peak', 'All Time Peak', 'Ref.'], axis= 1)
df
```

```python
df.drop_duplicates()
df.dropna()
df
```

```python
df = df.rename(columns= {'Actual gross': 'Total Earn', 'Adjusted gross (in 2022 dollars)': 'Profits (2022)',
                         'Average gross': 'Average Earning', 'Tour title': 'Tour Title'})
df
```

```python
df['Total Earn'] = df['Total Earn'].str.split(pat= '$').str[1]
df['Total Earn'] = df['Total Earn'].str.split(pat= '[').str[0]
df['Profits (2022)'] = df['Profits (2022)'].str.split(pat= '$').str[1]
df['Average Earning'] = df['Average Earning'].str.split(pat= '$').str[1]

df['Total Earn'] = df['Total Earn'].str.replace(',', '').astype('int')
df['Profits (2022)'] = df['Profits (2022)'].str.replace(',', '').astype('int')
df['Average Earning'] = df['Average Earning'].str.replace(',', '').astype('int')
df
```

```python
df['Tour Title'] = df['Tour Title'].str.split(pat= '†').str[0]
df['Tour Title'] = df['Tour Title'].str.split(pat= '‡').str[0]
df['Tour Title'] = df['Tour Title'].str.split(pat= '*').str[0]
df
```

```python
df['Start Year'] = df['Year(s)'].str.split(pat= '–').str[0].astype('int')
df['End Year'] = df['Year(s)'].str.split(pat= '–').str[1]
df['End Year'] = df['End Year'].fillna(df['Year(s)'].str.split(pat= '–').str[0]).astype('int')
df = df.drop(columns= 'Year(s)', axis= 1)
df
```

```python
plt.bar(df['Artist'], df['Average Earning'], color= 'orange')
plt.title('Total Earn by Artist')
plt.xlabel('Artist')
plt.ylabel('Total Earn ($ Million)')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
Artist_Profit = df.groupby('Artist')['Profits (2022)'].sum()
plt.pie(Artist_Profit, labels= Artist_Profit.index, autopct= '%0.1f%%', startangle= 90)
plt.title('Profit in 2022 by Artist')
```

```python
plt.plot(df['Start Year'], df['Average Earning'], marker= 'o', color= 'brown')
plt.title('Average Earning Over Time')
plt.xlabel('Start Year')
plt.ylabel('Avrage Earning ($ Million)')
plt.grid(True)
plt.tight_layout()
```

```python
plt.scatter(df['Shows'], df['Profits (2022)'], color= 'blue')
plt.title('Profit Vs Profit in 2022')
plt.xlabel('Number of Shows')
plt.ylabel('Profit in 2022 ($ Million)')
plt.grid(True)
plt.tight_layout()
```

```python
plt.hist(df['Shows'], bins= 10, color= 'pink', edgecolor= 'black')
plt.title('Distribution of Numbers of Shows')
plt.xlabel('Number Of Shows')
plt.ylabel('Frequency')
plt.tight_layout()
```
